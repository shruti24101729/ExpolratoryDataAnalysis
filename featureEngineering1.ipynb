```python
##feature scaling---standardization,normalization
#normalisation---minmax , robust
#standardisation is called z score normalisation formula:x(i)=(x(i)-mean(x))/sd(x)---used in K-means,kNN,PCA,gradient descent
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/Social_Network_Ads.csv')
df=df.iloc[:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.3,random_state=0)

x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler # used for standardisation
sc=StandardScaler()
sc.fit(x_train)
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)


sc.mean_
```

```python
x_train_sc
```

```python
x_train_sc=pd.DataFrame(x_train_sc,columns=x_train.columns)
x_test_sc=pd.DataFrame(x_test_sc,columns=x_test.columns)
x_train_sc
```

```python
x_train.describe(),x_train_sc.describe()
```

```python
import matplotlib.pyplot as plt
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(10,5))

ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title('Before Scaling')
ax2.scatter(x_train_sc['Age'],x_train_sc['EstimatedSalary'])
ax2.set_title('After Scaling')
```

```python
import seaborn as sns
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(10,5))
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)
ax1.set_title('Before Scaling')


sns.kdeplot(x_train_sc['Age'],ax=ax2)
sns.kdeplot(x_train_sc['EstimatedSalary'],ax=ax2)
ax2.set_title('After Scaling')

plt.show()
```

```python
fig,(ax1,ax2,ax3,ax4)=plt.subplots(ncols=4,figsize=(15,2))
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax3,color='r')
ax1.set_title(' Age Before Scaling')
ax3.set_title('EstimatedSalary Before Scaling')


sns.kdeplot(x_train_sc['Age'],ax=ax2)
sns.kdeplot(x_train_sc['EstimatedSalary'],ax=ax4,color='r')
ax2.set_title('Age After Scaling')
ax4.set_title('EstimatedSalary After Scaling')

plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
lr=LogisticRegression()
#before scaling
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#after scaling
lr.fit(x_train_sc,y_train)
y_pred=lr.predict(x_test_sc)
accuracy_score(y_test,y_pred)

```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
#before scaling
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
dt.fit(x_train_sc,y_train)
y_pred=dt.predict(x_test_sc)
accuracy_score(y_test,y_pred)
```

```python
df=pd.read_csv('/content/covid_toy.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('has_covid',axis=1),df['has_covid'],test_size=0.2,random_state=0)
x_train
```

```python
#Method I
si = SimpleImputer()
x_train_fever=si.fit_transform(x_train[['fever']])
x_test_fever=si.fit_transform(x_test[['fever']])
x_train_fever.shape,x_test_fever.shape
```

```python
oe=OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough=oe.fit_transform(x_train[['cough']])
x_test_cough=oe.fit_transform(x_test[['cough']])
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False)
x_train_gender_city=ohe.fit_transform(x_train[['gender','city']])
x_test_gender_city=ohe.fit_transform(x_test[['gender','city']])
```

```python
x_train_age=x_train.drop(columns=['gender','fever','cough','city']).values
x_test_age=x_test.drop(columns=['gender','fever','cough','city']).values
```

```python
x_train_transformed=np.concatenate((x_train_age,x_train_fever,x_train_gender_city,x_train_cough),axis=1)
x_test_transformed=np.concatenate((x_test_age,x_test_fever,x_test_gender_city,x_test_cough),axis=1)
x_train_transformed=pd.DataFrame(x_train_transformed)
x_test_transformed=pd.DataFrame(x_test_transformed)
x_train_transformed,x_train
```

```python
#method 2 using column transformer
from sklearn.compose import ColumnTransformer
transformer=ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
     ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
     ('tnf3',OneHotEncoder(drop='first',sparse_output=False),['gender','city'])
      ],remainder='passthrough')
```

```python
transformer.fit_transform(x_train)
```
