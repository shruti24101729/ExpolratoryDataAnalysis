```python
batters=pd.read_csv('/content/batter.csv')
batters.head()
```

```python
# we will take only 25 batsman randomly

sample_df=batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(15,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

#annotations can be helpful in identifing players

for i in range(25):
  plt.annotate(sample_df['batter'].values[i],(sample_df['avg'].values[i],sample_df['strike_rate'].values[i]))
```

```python
# looking for strike rate>130
plt.figure(figsize=(15,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red',linestyle='dashed')
plt.axvline(30,color='red',linestyle='dashed')

for i in range(25):
  plt.annotate(sample_df['batter'].values[i],(sample_df['avg'].values[i],sample_df['strike_rate'].values[i]))
```

```python
############# subplots ######################

batters.head()
```

```python
## another way of plotting graph

fig,ax=plt.subplots(figsize=(15,7))
ax.scatter(batters['avg'],batters['strike_rate'],color='red',marker='+')
ax.set_title('Avg vs SR')
ax.set_xlabel('Avg')
ax.set_ylabel('SR')

fig.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(15,7),sharex=True) #figure object, axis object

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0].set_title('Avg vs SR')
ax[0].set_xlabel('Avg')
ax[0].set_ylabel('SR')

ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_title('Avg vs Runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')

fig.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10)) #figure object, axis object

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])


ax[1,0].hist(batters['strike_rate'],color='red')
ax[1,1].hist(batters['runs'])

fig.show()
```

```python
########### one more way to plot


fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')
ax2=fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'],batters['runs'])
ax3=fig.add_subplot(2,2,3)
ax3.hist(batters['strike_rate'],color='red')
ax4=fig.add_subplot(2,2,4)
ax4.hist(batters['runs'])

fig.show()
```

```python
###### 3D Scatter plots


fig=plt.figure()
ax=fig.add_subplot(projection='3d')

ax.scatter3D(batters['avg'],batters['strike_rate'],batters['runs'])
ax.set_xlabel('Avg')
ax.set_ylabel('SR')
ax.set_zlabel('Runs')

fig.show()
```

```python
### 3D line plot

x=[0,1,5]
y=[0,10,13]
z=[0,13,20]

fig=plt.figure()
ax=fig.add_subplot(projection='3d')

#ax.scatter3D(batters['avg'],batters['strike_rate'],batters['runs'])
ax.scatter3D(x,y,z,s=[100,100,100])
ax.plot3D(x,y,z,color='red')

fig.show()
```

```python
## 3D Surface plot

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)


xx,yy=np.meshgrid(x,y)
#z=xx**2+yy**2
z=np.sin(xx)+np.cos(yy)

fig=plt.figure(figsize=(15,7))
ax=fig.add_subplot(projection='3d')

ax.plot_surface(xx,yy,z,cmap='plasma')

fig.show()
```

```python
##  contour plot ############
# it represents 3D graph into 2D


x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)


xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
#z=np.sin(xx)+np.cos(yy)

fig=plt.figure()

ax=fig.add_subplot(1,2,1)
ax.contour(xx,yy,z,cmap='plasma')

ax1=fig.add_subplot(1,2,2,projection='3d')
#ax1=fig.add_subplot(projection='3d')
ax1.plot_surface(xx,yy,z,cmap='plasma')

fig.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.contourf(xx,yy,z,cmap='plasma')## filled contour
fig.show()
```

**Seaborn**

Two functions- figure level, axis level

Seaborn has relational plot(scatterplot,lineplot), distribution plot also called displot(histogram, kdeplot,rugplot), categorical plot(bar plot,countplot,boxplot,violinplot,swarmplot), matrix plot(heatmap,clustermap),regression plot,multiplots(joint plot,pairplot)

Seaborn does not support animations

seaborn do not have 3d visualisation , pie chart , subplots. So we need matplotlib

```python
#relational plot
# - statstical relation
# -bivariate analysis

import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df=sns.load_dataset('tips')
df
```

```python
#scatter plot-> axes level function
sns.scatterplot(x='total_bill',y='tip',data=df,color='red',hue='sex',style='smoker',size='size')
```

```python
# relplot(relational plot ) is a figure level function-- figure shape is square shape
#but axes level is quite rectangular
#use figure level function over axes level function  because from figure level you can plot many graphs
```

figure level function catplot

```python
#axes level function
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False,hue='sex')
```

```python
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',kind='strip',jitter=1,hue='sex')#default jitter=1
#jittering is done for better readibility
```

```python
#swarmplot
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
#swarm is basically better representation of strip for better data distribution
#swarm works only for small datasets but strip works on all
```

A boxplot is a standardised way of displaying the distribution of data based on a five number summary("minimum",first quartile[Q1],median,third quartile[Q3], and "maximum". It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewd.
minimum=Q1-1.5*IQR, maximum=Q3+1.5*IQR , IQR is interquartile range,IQR=Q3-Q1
